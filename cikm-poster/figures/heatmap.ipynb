```python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt


from matplotlib import rc
rc('font', **{'family': 'serif', 'sans-serif': ['cm']})
rc('text', usetex=True)
rc('axes', **{'titlesize': 'large'})
plt.rcParams['text.latex.preamble'] = """
\\usepackage{libertine}
\\usepackage[libertine]{newtxmath}
"""

plt.rcParams['figure.figsize'] = (2.7, 2.1)
plt.rcParams['figure.dpi'] = 240
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
```

```python
image = np.array([[0.03309322, 0.02489234, 0.02479361, 0.02474955, 0.02465459],
       [0.02489234, 0.03569873, 0.0242462 , 0.02679288, 0.02450668],
       [0.02479361, 0.02424619, 0.03422668, 0.02688449, 0.        ],
       [0.02474955, 0.02679288, 0.02688449, 0.03488097, 0.02346804],
       [0.02465459, 0.02450668, 0.        , 0.        , 0.03462299]])
```

```python
sns.set(font_scale=1.5)

h = sns.heatmap(data=image,square=True,xticklabels=False, yticklabels=False,cmap=plt.get_cmap('Oranges'),cbar=False,
               linewidths=2, linecolor='w')

cb = h.figure.colorbar(h.collections[0])
cb.set_ticks([0,0.03])
plt.savefig('visu-2.pdf', bbox_inches='tight')
```
